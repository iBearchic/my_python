```python
currencies = response.json()
currencies
```

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/it_new.csv')
df.head(10)
```

```python
df['curenc'].unique()
```

```python
df['curenc']=df['curenc'].replace({'zł':'PLN', 'Kč':'CZK', '$':'USD', 'lev':'BGN', 'Ft':'HUF', '€':'EUR'})
df.head(10)
```

```python
def exchange(row):
    if (row['curenc']!='₽'):
        return "{:.15f}".format(row['col']*currencies['Valute'][row['curenc']]['Value'])
    else:
        return row['col']
```

```python
df['col']=df.apply(exchange,axis=1)
df
```
