```python
import pandas as pd

url = "https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/deputat.csv"
df = pd.read_csv(url)
df
```

```python
import datetime
from datetime import datetime, timedelta

def convert_to_datetime(row):
    return datetime.strptime(row['birthdate'], '%d.%m.%Y')
```

```python
df['birthdate']=df.apply(convert_to_datetime, axis=1)
```

```python
import time
def make_unix_time(row):
    return time.mktime(row['birthdate'].timetuple())
```

```python
df['Возраст']=list(map(lambda x: 2020-x.year,df['birthdate']))
```
